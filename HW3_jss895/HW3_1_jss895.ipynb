# Interpretation:

This histogram shows how the means are distributed, (it looks like a Gaussian with mean 12). Which is what we set the distributions to do, confirming the central limit theorem.

```python
mean = 12
std = 10
lambdaa = 12
k = 12
n = 24
p = 0.5
scale = 12

means=[]


np.random.seed(14)

for i in range (100):
    randsize = np.random.randint(10, 2000)
    sample = np.random.exponential (scale,randsize)
    means.append(sample.mean())

for i in range (100):
    randsize = np.random.randint(10, 2000)
    sample = np.random.normal (mean,std,randsize)
    means.append(sample.mean())

for i in range (100):
    randsize = np.random.randint(10, 2000)
    sample = np.random.poisson (lambdaa,randsize)
    means.append(sample.mean())
    
for i in range (100):
    randsize = np.random.randint(10, 2000)
    sample = np.random.chisquare (k,randsize)
    means.append(sample.mean())
    
for i in range (100):
    randsize = np.random.randint(10, 2000)
    sample = np.random.binomial (n,p,randsize)
    means.append(sample.mean())
    
    
    
sortmeans = sorted(means)
fit = stats.norm.pdf(sortmeans, np.mean(sortmeans), np.std(sortmeans))
plt.plot(sortmeans, fit)

plt.hist(means, bins=100, normed=True);

plt.xlabel("Mean", fontsize=12)
plt.ylabel("Mean count Normalized", fontsize=12)
plt.title("Mean Distribution", weight='bold', fontsize=14)
xlim (8,16)

```
